```python
import matplotlib.pyplot as plt
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]
plt.plot(x,cxMarks, marker="v", c="r", label="CX", ls="--")
plt.plot(x,cyMarks, marker="o", c="b", label="CY", ls="-.")
plt.annotate("Lowest CPI", xy=[8,1])
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.legend()
plt.show()
```

### 04) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
dept=["CE","Civil","Mech","Diploma","BCA"]
students=[1000,300,250,500,400]
plt.pie(students,labels=dept,colors=["r","b","k","y","g"], radius=0.9)
plt.title("Departments")
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
plt.bar(dept,students,width=0.4)
plt.xlabel("Departments")
plt.ylabel("Students")
plt.show()
```

### 06) WAP to demonstrate the use of Scatter Plot.

```python
import matplotlib.pyplot as plt
import random
random.seed(10)
x=[random.random() for i in range(50)]
y=[random.random() for i in range(50)]
z=[random.random() for i in range(50)]
plt.scatter(x,y,cmap="plasma",c=z)
plt.colorbar()
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import matplotlib.pyplot as plt
import random
random.seed(7)
x=[random.random() for i in range(50)]
plt.hist(x,edgecolor="k",bins=5,color="g",histtype="stepfilled")
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
dept=["CE","Civil","Mech","Diploma","BCA"]
s1=[1000,300,250,500,400]
b=plt.bar(dept,s1,width=0.3, color=["r"])
for i in b:
    plt.text(i.get_x(),i.get_height()+10,f"{i.get_height()}")
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import matplotlib.pyplot as plt
import random
random.seed(10)
x=[random.random() for i in range(100)]
y=[random.random() for i in range(100)]
z=[random.random() for i in range(100)]
plt.scatter(x,y,cmap="viridis",c=z)
plt.colorbar()
plt.show()
```

### 10) WAP to create a Box Plot.

```python
import matplotlib.pyplot as plt
import random
random.seed(10)
time=[random.randint(2,120) for i in range(100)]
plt.figure(figsize=(8,5))
plt.boxplot(time,patch_artist=True, boxprops=dict(facecolor="skyblue"))
plt.title("Box Plot with Various Properties")
plt.xlabel("Data Distribution")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
