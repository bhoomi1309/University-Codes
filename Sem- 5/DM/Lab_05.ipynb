### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.

```python
#Min Max
df['Age'] = df['Age'].interpolate()
min_age = df['Age'].min()
max_age = df['Age'].max()
df['Age_Scaled'] = (df['Age'] - min_age) / (max_age - min_age)
df['Age_Scaled']
```

```python
#Decimal Scaling
j = len(str(abs(df['Age'].max())))
df['Age_Scaled'] = df['Age'] / (10 ** j)
df['Age_Scaled']
```

```python
#Z Score
mean_age = df['Age'].mean()
std_age = df['Age'].std()
df['Age_Scaled'] = (df['Age'] - mean_age) / std_age
df['Age_Scaled']
```
